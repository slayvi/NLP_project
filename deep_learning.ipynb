# Sentiment Analysis with Deep Learning

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.layers import TextVectorization
import tensorflow as tf

tf.random.set_seed(42)
```

Loading training / validation set (preprocessed):

```python
batch_size = 1024
seed = 42
train_ds = tf.keras.utils.text_dataset_from_directory("./dataset/train/pre_final", batch_size=batch_size, validation_split=0.2, subset="training", seed=seed)
val_ds = tf.keras.utils.text_dataset_from_directory("./dataset/train/pre_final", batch_size=batch_size, validation_split=0.2, subset="validation", seed=seed)
```

```python
print("Label 0: ", train_ds.class_names[0])
print("Label 1: ", val_ds.class_names[1])
```

Fetching some examples for testing implementation code:

```python
for text_batch, label_batch in train_ds.take(1):
    for i in range(5):
        print(label_batch[i].numpy(), text_batch.numpy()[i])
```

Optimize tf:

```python
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

## Build the Model

```python
vocab_size = 15000 
sequence_length = 500
embedding_dim = 8 
```

Input Layer:

```python
vectorize_layer = TextVectorization(
    standardize = None, 
    max_tokens = vocab_size, 
    ngrams = (1,2),  
    output_mode = "int", 
    output_sequence_length = sequence_length
)
```

Map and adapt to build the vocabulary:

```python
text_ds = train_ds.map(lambda x, y:x)
vectorize_layer.adapt(text_ds)
```

Build the architecture:

```python
model = Sequential([
    vectorize_layer, 
    Embedding(vocab_size, embedding_dim, name="embedding"), 
    GlobalAveragePooling1D(),    
    Dense(16, activation="relu"), 
    Dense(1)
])
```

Optimizing training process and plot graphics to tensorboard:

```python
import datetime 
log_dir = "./logs" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S") + "final"
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
tensorboard_callback_early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=3,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=15
)
```

Compile the model:

```python
model.compile(optimizer="adam", 
              loss = tf.keras.losses.BinaryCrossentropy(from_logits=True), 
              metrics = ["accuracy"])
```

And finally fit it: 

```python
model.fit(
    train_ds,
    validation_data = val_ds,
    epochs = 100, 
    callbacks=[tensorboard_callback, tensorboard_callback_early_stopping]
)
```

Show model summary:

```python
model.summary()
```

Save model:

```python
model.save('./models/model_NN_final.tf')
```

```python
weights = model.get_layer("embedding").get_weights()[0]
vocab = vectorize_layer.get_vocabulary()
```

```python
import io 

out_v = io.open("./models/vectors_NN_final.tsv", "w", encoding="utf-8")
out_m = io.open("./models/metadata_NN_final.tsv", "w", encoding="utf-8")

for index, word in enumerate(vocab):
    if index == 0:
        continue  # skip 0, its padding

    else:
        vec = weights[index]
        out_v.write("\t".join([str(x) for x in vec]) + "\n")
        out_m.write(word + "\n")

out_v.close()
out_m.close() 
```

# Testing

Load preprocessed test set: 

```python
test_ds = tf.keras.utils.text_dataset_from_directory("./dataset/test/pre_final")
```

Evaluate model on test set:

```python
results = model.evaluate(test_ds)

for name, value in zip(model.metrics_names, results):
  print("%s: %.3f" % (name, value))
```

Preprocessing the input:

```python
import spacy
nlp = spacy.load("en_core_web_sm")
import re 
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
stop_words.remove("not")
stop_words.remove("nor")
stop_words.remove("no")
stop_words.remove("again")
add_stopwords  = set(["movie", "film", "one", "the", "scene", "this", "story", "would", "really", "and", "also", ])
stop_words = stop_words.union(add_stopwords)

def remove_html(text):
    text = re.sub(r"<[\w]+ />", " ", text)
    text = re.sub("n't", " not", text)
    return text 

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\b\w{1,1}\b", " ", text)                                                     # remove single characters
    text = re.sub(r"[^a-z]", " ", text)                                                          # remove everything which aren't letters
    text = re.sub(r"[\s]+", " ", text)                                                           # remove too many whitespaces
    return text

def process_and_filter_non_entities(text):
    doc = nlp(text)
    non_entity_lemmas = [token.lemma_ for token in doc if token.ent_type_ != "PERSON"]
    non_entity_lemmas = [token for token in non_entity_lemmas if token.lower() not in stop_words]
    text = " ".join(non_entity_lemmas) 
    return text
```

# Test manually to plot confusion matrix:

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
data = pd.read_csv("./dataset/testing_set_preprocessed.csv")
data
```

```python
features = data[['sentiment', 'review']] 
```

Retrieving true and predicted labels for plotting CM:


```python
pred_labels = []
true_labels = []

for i in range(len(features)):
    print(i)
    temp = model.predict([features["review"][i]])
    pred_labels.append(temp)
    true_labels.append(features["sentiment"][i])
```

Convert predictions into binary and store into list:

```python
pred_label_binary = []

for idx in range(len(pred_labels)):
    if pred_labels[idx] < 0:
        pred_label_binary.append(0)
    else:
        pred_label_binary.append(1)
```

```python
true_label_int = []

for i in range(len(true_labels)):
    if true_labels[i] == "pos":
        true_label_int.append(1)
    else:
        true_label_int.append(0)
```

```python
y_true_classes = ["neg", "pos"]
class_to_index = {class_name: i for i, class_name in enumerate(y_true_classes)}
y_true_labels = [x for x in true_label_int]
pred_labels = [pred for pred in pred_label_binary]

# Create confusion matrix
conf_mtx = confusion_matrix(y_true_labels, pred_labels)
conf_mtx_normalized = confusion_matrix(y_true_labels, pred_labels, normalize="true")

# Set up subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot raw counts
sns.heatmap(conf_mtx, annot=True, fmt='g', cmap='Blues', xticklabels=y_true_classes, yticklabels=y_true_classes, ax=axs[0], cbar=False,   annot_kws={"size": 18})
axs[0].set_xlabel('Predicted')
axs[0].set_ylabel('True')
axs[0].set_title('Confusion Matrix Neural Network', fontsize=20)

# Plot normalized values
sns.heatmap(conf_mtx_normalized, annot=True, fmt='.2f', cmap='Blues', xticklabels=y_true_classes, yticklabels=y_true_classes, ax=axs[1], cbar=False,   annot_kws={"size": 18}) 
axs[1].set_xlabel('Predicted')
axs[1].set_ylabel('True')
axs[1].set_title('Normalized Confusion Matrix Neural Network', fontsize=20)

plt.savefig("./results/CM_embed_raw_norm.png", dpi=360)
plt.show()
```

Print all metrics:

```python
print("Accuracy: ", accuracy_score(true_label_int, pred_label_binary))
print("Precision: ", precision_score(true_label_int, pred_label_binary))
print("Recall: ", recall_score(true_label_int, pred_label_binary))
print("F1-score: ", f1_score(true_label_int, pred_label_binary))
```
